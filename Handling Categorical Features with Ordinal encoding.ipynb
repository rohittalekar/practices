## Ordinal Number Encoding

```python
import datetime
```

```python
today_date = datetime.datetime.today()
```

```python
today_date
```

```python
days = [today_date - datetime.timedelta(x) for x in range(15)]
```

```python
import pandas as pd
df = pd.DataFrame(days)
df.columns =["day"]
```

```python
df['weekday'] = df['day'].dt.day_name()
```

```python
weekday_dictonary = {"Monday" :1, "Tuesday":2, "Wednesday":3, "Thursday":4, "Friday":5, "Saturday":6, "Sunday":7}
```

```python
df['weekday_ordinal'] = df['weekday'].map(weekday_dictonary)
```

```python
df.head()
```

## Target Guided Ordinal Encoding

```python
import pandas as pd

df = pd.read_csv('TatanicDataSet.csv',usecols=['Cabin','Survived'])
```

```python
df['Cabin'] =df['Cabin'].fillna('Not exits')
```

```python
df['Cabin_cat'] = df['Cabin'].astype(str).str[0]
```

```python
df.Cabin_cat.unique()
```

```python
df.groupby('Cabin_cat')['Survived'].mean()
```

```python
df.groupby('Cabin_cat')['Survived'].mean().sort_values().index
```

```python
ordinal_lables = df.groupby('Cabin_cat')['Survived'].mean().sort_values().index
ordinal_lables
```

```python
ordinal_lables1 = {k:i for i,k in enumerate(ordinal_lables,0)}
```

```python
ordinal_lables1
```

```python
df['Cabin_cat'] = df['Cabin_cat'].map(ordinal_lables1)
```

```python
df.head()
```

## Mean Encoding

```python
mean_ordinal = df.groupby('Cabin_cat')['Survived'].mean().to_dict()
```

```python
df['Cabin_cat_mean_ordinal'] = df['Cabin_cat'].map(mean_ordinal)
```
