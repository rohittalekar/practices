```python
#Remove outliers using IQR techniques

import pandas as pd
```

```python
df = pd.read_csv('sample_data/TitanicDataSet.csv')
```

```python
import seaborn as sns

sns.distplot(df['Age'].fillna(100))
```

```python
##Gausian Distribution
```

```python
df.Age.hist(bins=50)
```

```python
df.boxplot(column="Age")
```

```python
## We will calculate the upper and lower boundaries values

upper_boundary = df["Age"].mean() + 3 * df["Age"].std()
upper_boundary
```

```python
lower_boundary = df["Age"].mean() - 3 * df["Age"].std()
lower_boundary
```

```python
## Calculate IQR
```

```python
IQR = df.Age.quantile(0.75) - df.Age.quantile(0.25)
IQR
```

```python
lower_bridge = df["Age"].quantile(0.25) - (IQR*1.5)
lower_bridge
```

```python
upper_bridge = df["Age"].quantile(0.25) - (IQR*1.5)
upper_bridge
```

```python
#Extreme Outliers
lower_bridge_extreme= df["Age"].quantile(0.25) - (IQR*3)
upper_bridge_extreme= df["Age"].quantile(0.75) + (IQR*3)
```
