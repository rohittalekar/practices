```python
#Gaussian Transformation
```

```python
import pandas as pd
df = pd.read_csv('TitanicDataSet.csv', usecols=['Age','Fare','Survived'])
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
# Check features are normally distributed or not with the QQ plot
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stat
import pylab
```

```python
def plot_data(df, feature):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  df[feature].hist()
  plt.subplot(1,2,2)
  stat.probplot(df[feature],dist='norm',plot=pylab)
  plt.show()
```

```python
plot_data(df,'Age')
```

```python
#Logrithmic Transformation
```

```python
import numpy as np

df['Age_log'] = np.log(df['Age'])
plot_data(df,'Age_log')

##We cannot use Logrithmic transformation as it is not in straight line
```

```python
#Resiprocal Transformation
```

```python
df['Age_resiprocal'] = 1/df.Age
plot_data(df, 'Age_resiprocal')
```

```python
#Square root transformation
```

```python
df['Age_square'] = df.Age**(1/2)
plot_data(df,'Age_square')
```

```python
##Exponential Transformation
```

```python
df['Age_Exponential'] = df.Age**(1/1.2)
plot_data(df, 'Age_Exponential')
```

```python
#Box-Cox
```

```python
df['Age_boxcox'], parameter = stat.boxcox(df['Age'])
```

```python
plot_data(df,'Age_boxcox')
```
