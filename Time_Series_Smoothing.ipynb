```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
mpl.rcParams['figure.figsize'] = 20,5
```

```python
electricitydataset = pd.read_csv('Electric_Production.csv', header=0,index_col=0)
```

```python
## Moving Average
```

```python
plt.plot(electricitydataset[1:50]['IPG2211A2N'])
plt.xticks(rotation=30)
plt.show()
#There is seasonality in the graph as well as the trend.
```

```python
##Rolling Average Transform

rollingseries = electricitydataset[1:50].rolling(window=10)

rollingmean = rollingseries.mean()

print(rollingmean.head(10)) ## 9 values will be blank as MA = 10
```
